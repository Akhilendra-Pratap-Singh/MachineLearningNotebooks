```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
encoded_columns = ohe.fit_transform(df[['ocean_proximity']])
ocean_encoded = pd.DataFrame(encoded_columns, columns= ohe.get_feature_names_out(['ocean_proximity']))
```

```python
df = pd.concat([df,ocean_encoded],axis = 1)
```

```python
df.head()
```

```python
df.drop('ocean_proximity',axis = 1,inplace = True)
```

```python
df.info()
```

```python
X = df.drop('median_house_value',axis = 1)
y = df['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
# from sklearn.linear_model import LinearRegression
# regression =  LinearRegression()
# regression.fit(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
regression = RandomForestRegressor()
regression.fit(X_train,y_train)
```

```python
# from sklearn.tree import DecisionTreeRegressor
# regression = DecisionTreeRegressor()
# regression.fit(X_train,y_train)
```

```python
# from sklearn.svm import SVR
# regression = SVR(kernel='rbf')
# regression.fit(X,y)
```

```python
y_pred = regression.predict(X_test)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Create a figure
plt.figure(figsize=(10, 6))

# Scatter plot of actual and predicted values
# Use the index of y_test for the x-coordinates and its values for y-coordinates
plt.scatter(y_test.index, y_test.values, color='red', label='Actual values', alpha=0.6)
plt.scatter(y_test.index, y_pred, color='blue', label='Predicted values', alpha=0.6)  # Use the same index for y_pred

# Line of perfect prediction
min_val = min(min(y_test), min(y_pred))
max_val = max(max(y_test), max(y_pred))
plt.plot([min_val, max_val], [min_val, max_val], color='green', linestyle='--', label='Perfect prediction')

# Labels and title
plt.xlabel('Sample Index')
plt.ylabel('Values')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = regression.predict(X_test)

# Calculate regression metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```
